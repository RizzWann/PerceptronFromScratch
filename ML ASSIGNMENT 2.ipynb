```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("E:MSDS/MS-2/ML/iris.csv")
df 
```

```python
df = df.iloc[0:-1,0:-10]
df
```

```python
df_shuffled = df.sample(150,random_state=42)
df_shuffled.head()
```

```python
df_shuffled.isna().sum()
```

```python
X = df_shuffled.iloc[:,:-1].values
X
```

```python
y = df_shuffled.Species
y
```

```python
X_train = X[0:75,:]
X_test = X[75:,:]
y_train = np.array([y[0:75]])
y_test = np.array([y[75:]])
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
def gradient_descent(X, y_true, epochs, learning_rate = 0.01):

    
    w = 0.1*np.random.randn(1,X.shape[1])
    b = - 0.5#bias

    n = X.shape[0] # number of rows in X
    
    cost_list = []
    epoch_list = []
  
    for i in range(epochs):    
        # (1,4)(4,75) = (1,75)
        y_predicted = np.dot(w, X.T) + b

        diff = y_true - y_predicted
        # (1,75)(75,4)
        w_grad = -(1/n)*(diff.dot(X))
        b_grad = -(1/n)*np.sum(diff)
      
        w = w - learning_rate * w_grad
        b = b - learning_rate * b_grad
        
        cost = (np.mean(np.square(diff)))/2 # MSE (Mean Squared Error)
        
        if i%1000==0:
            print('iterations#',i)
            print(cost)
            cost_list.append(cost)
            epoch_list.append(i)
        
    return w, b, cost, cost_list, epoch_list
```

```python
w, b, cost, cost_list, epoch_list=gradient_descent(X_train, y_train, 204800, learning_rate = 0.01)
w,b
```

```python
plt.xlabel("epoch")
plt.ylabel("cost")
plt.plot(epoch_list,cost_list)
```

```python
def accuracy_wdout_limit(w,b,X_test):
    count=0
    n = len(X_test)
    for i in range(n):
        pred = np.dot(w,X_test[i]) + b
        if pred[0] == y_test[0][i]:
            print("matched")
            count += 1
    return count/n,count      
```

```python
accuracy_wdout_limit(w,b,X_test)
```

```python
def accuracy_limit(w,b,X_test):
    count=0
    n = len(X_test)
    for i in range(n):
        pred = np.dot(w,X_test[i]) + b
        if pred <= 1.5:
            pred = 1.0
        elif pred < 1.5 and pred <= 2.5:
            pred = 2.0
        elif pred>2.5:
            pred = 3.0
        if pred == y_test[0][i]:
            count += 1
    return count/n,count
```

```python
accuracy_limit(w,b,X_test)
```
